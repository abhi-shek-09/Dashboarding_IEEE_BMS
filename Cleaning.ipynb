```python
def fill_missing_profits(row):
    if pd.notnull(row['DifferenceBetweenActualandTargetProfit']):
        if pd.isnull(row['Profit']) and pd.notnull(row['Target_profit']):
            return row['DifferenceBetweenActualandTargetProfit'] - row['Target_profit']
        elif pd.isnull(row['Target_profit']) and pd.notnull(row['Profit']):
            return row['DifferenceBetweenActualandTargetProfit'] + row['Profit']
    return pd.Series([row['Profit'], row['Target_profit']])

# Apply the function to fill missing values
df[['Profit', 'Target_profit']] = df.apply(fill_missing_profits, axis=1)
```

```python
df.isnull().sum()
```

```python
df['Target_margin'].fillna(df['Target_margin'].mean(), inplace=True)
```

```python
df['Target_margin'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
def calculate_target_sales(row):
    if pd.notnull(row['Target_cogs']) and pd.notnull(row['Target_margin']):
        if row['Target_margin'] != 0:
            return row['Target_cogs'] / (1 - row['Target_margin'] / 100)
        else:
            return np.nan
    else:
        return np.nan

df.loc[df['Target_sales'].isnull(), 'Target_sales'] = df[df['Target_sales'].isnull()].apply(calculate_target_sales, axis=1)
```

```python
df.isnull().sum()
```

```python
def calculate_target_cogs(row):
    if pd.notnull(row['Sales']):
        cogs_percentage = 70
        return row['Sales'] * (cogs_percentage / 100)
    else:
        return np.nan

df.loc[df['Target_cogs'].isnull(), 'Target_cogs'] = df[df['Target_cogs'].isnull()].apply(calculate_target_cogs, axis=1)
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Cogs'])
```
